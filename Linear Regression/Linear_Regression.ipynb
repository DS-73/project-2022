```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split  

bodyFat = pd.read_csv('bodyfat.csv')
# x = Independent Variable
# y = Dependent Variable
x = bodyFat.iloc[:,0].values
y = bodyFat.iloc[:,1].values
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.30)
x_train = x_train.reshape(-1,1)
x_test = x_test.reshape(-1,1)
regressor = LinearRegression()

model = regressor.fit(x_train, y_train)
print("Model Score: " + model.score(x_train, y_train))

slope = regressor.coef_
print("Slope of line: ",slope)
intercept = regressor.intercept_
print("Intercept of line: ",intercept)
print("Equation of Line: y = ",slope,"x + " ,intercept)

y_pred = regressor.predict(x_test)
plt.plot(x_test, y_pred, color='yellow', linewidth=0.3)

plt.scatter(x_train, y_train, color="#127369")   
plt.title("Body Fat (Training Dataset)")  
plt.xlabel("Density")  
plt.ylabel("Body Fat")  
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split  

bodyFat = pd.read_csv('bodyfat.csv')
# x = Independent Variable
# y = Dependent Variable
x = bodyFat.iloc[:,0].values
y = bodyFat.iloc[:,1].values

n = np.size(bodyFat, 0)
Y = np.mean(y)
X = np.mean(x)
sumY = np.sum(y)
sumX = np.sum(x)
sumXX = np.sum(x**2)
sumXY = np.sum(x*y)
A = ((n*sumXY) - (sumX * sumY)) / ((n*sumXX) - (sumX)**2)
var = A / n
B = Y - var*sumX

print("Slope of line: ",A)
print("Intercept of line: ",B)
print("Equation of Line: y = ",A,"x + " ,B)

plt.plot(x, y, color='yellow', linewidth=0.3)

plt.scatter(x, y, color="#127369")   
plt.title("Body Fat (Training Dataset)")  
plt.xlabel("Density")  
plt.ylabel("Body Fat")  
plt.show()
```
